```python
# Import necessary modules
import numpy as np
import matplotlib.pyplot as plt

# Declare a list and a NumPy array
my_list = [1, 2, 3, 4, 5]
my_array = np.array([10, 20, 30, 40, 50])

# Define a basic Python function
def multiply_elements(input_list, multiplier):
    """
    This function multiplies each element of the input_list by the multiplier.
    Args:
    - input_list: List of numbers to be multiplied.
    - multiplier: Number to multiply each element by.
    
    Returns:
    - A list with the multiplied elements.
    """
    return [x * multiplier for x in input_list]

# Call the function with the list
multiplied_list = multiply_elements(my_list, 10)

# Print the results
print("Original list:", my_list)
print("Multiplied list:", multiplied_list)

# Create a simple plot using Matplotlib
plt.figure(figsize=(8, 6))
plt.plot(my_list, my_array, label='List vs Array', color='blue', marker='o')
plt.xlabel('List Values')
plt.ylabel('Array Values')
plt.title('Simple Plot of List vs Array')
plt.legend()

# Save the figure as a PNG
plt.savefig('simple_plot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the angles and radii for the star points
angles = np.linspace(0, 2 * np.pi, 11)  # 10 points for a 5-point star, plus closing point
radii = np.array([1, 0.4] * 5 + [1])  # Alternate between outer and inner points, closing with [1]

# Get x and y coordinates for the star shape
x = radii * np.cos(angles)
y = radii * np.sin(angles)

# Create the plot
plt.figure(figsize=(6, 6))
plt.plot(x, y, color='gold', marker='o', markersize=8, linestyle='-', linewidth=2)

# Formatting the plot to make it look more like a star
plt.gca().set_aspect('equal')  # Keep the aspect ratio equal to make the star symmetrical
plt.title('5-Point Star in Matplotlib', fontsize=14)
plt.grid(False)  # Turn off the grid
plt.axis('off')  # Turn off the axis for a cleaner look

# Save and display the figure
plt.savefig('star_plot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon
import random

# Function to create random polygons to simulate ice floes
def random_polygon(center, num_sides=6, scale=1):
    angle_step = 2 * np.pi / num_sides
    angles = np.arange(0, 2 * np.pi, angle_step) + np.random.uniform(0, angle_step, size=num_sides)
    radii = scale * (0.7 + 0.3 * np.random.rand(num_sides))  # Random variation in radius
    points = [(center[0] + r * np.cos(a), center[1] + r * np.sin(a)) for r, a in zip(radii, angles)]
    return points

# Set up the plot
fig, ax = plt.subplots(figsize=(8, 8))

# Ocean background (light blue)
ax.set_facecolor('lightblue')

# Generate random polygons representing ice floes with color variation
for _ in range(30):  # Create 30 ice floes
    center = np.random.uniform(-5, 5, size=2)
    num_sides = random.randint(5, 8)  # Random number of sides for each floe
    scale = random.uniform(0.5, 1.5)  # Random scale for size variation
    
    # Generate a random color from shades of white, light blue, and gray
    ice_color = (random.uniform(0.8, 1), random.uniform(0.9, 1), random.uniform(0.9, 1))  # Lighter colors for ice
    
    ice_polygon = random_polygon(center, num_sides=num_sides, scale=scale)
    
    # Create and add the polygon to the plot with varying colors
    polygon = Polygon(ice_polygon, closed=True, facecolor=ice_color, edgecolor='lightgray', lw=1)
    ax.add_patch(polygon)

# Adjust plot settings
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_aspect('equal')
plt.title('Colored Sea Ice Representation with Polygon Shapes', fontsize=14)
plt.axis('off')  # Turn off the axis for a cleaner look

# Save and display the plot
plt.savefig('colored_polygon_sea_ice_plot.png')
plt.show()
```
